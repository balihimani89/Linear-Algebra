Demonstrate in python and explain in your notebook by taking an example of a dataset that why you would or would not benefit from adding such kind of feature.
```python
import pandas as pd
```

```python
data = {'Student': ['A','B','C','D','E'],
        'Science': [35,45,10,15,20],'Math':[50,13,56,78,89],'Hindi':[66,77,88,99,44]}
```

```python
df = pd.DataFrame(data)
df
```

Criteria for admission in next class: pass in all 3 subjects and Marks in each subject>=35, pass

```python
df['Science_result'] = df['Science'].apply(lambda x: 'Pass' if x >= 35 else 'fail')
df['Math_result'] = df['Math'].apply(lambda x: 'Pass' if x >= 35 else 'fail')
df['Hindi_result'] = df['Hindi'].apply(lambda x: 'Pass' if x >= 35 else 'fail')
df
```

```python
df.loc[(df['Science_result'] == 'Pass') | (df['Math_result'] == 'Pass')|(df['Hindi_result'] == 'Pass'), 'Overall_Result'] = 'Promote to next Class' 
df.loc[(df['Science_result'] != 'Pass') | (df['Math_result'] != 'Pass')|(df['Hindi_result'] != 'Pass'), 'Overall_Result'] = 'Not Promoted'
df
```

if we combine any two columns and than try to predict the overall result, the overall results differs.
lets see

```python
df['Science_Math__Hindi_combine']=(df['Science']+df['Math']+df['Hindi'])/3
```

```python
df
```

```python
df['SMH_combine_result'] = df['Science_Math__Hindi_combine'].apply(lambda x: 'Pass' if x >= 35 else 'fail')
df
```

```python
df['Overall_combine_Result'] = df['SMH_combine_result'].apply(lambda x: 'Promote to next Class' if x == 'Pass' else 'Not Promoted')   
```

```python
df
```

```python
df[['Student','Overall_Result','Overall_combine_Result']]
```

Both results are different,would not benefit by linear combining of all the other feature columns.
